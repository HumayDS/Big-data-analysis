```python
import seaborn as sns
import pandas as pd
df = sns.load_dataset('titanic')
print(df.isnull().sum())
print(df.info())
```

```python
print(df['survived'].value_counts(normalize = True)) # Category
print(df['pclass'].value_counts()) # Category ordered
print(df['sex'].value_counts())  #Category
print(df['age'].value_counts()) #numeric int
print(df['sibsp'].value_counts()) #category
print(df['parch'].value_counts()) #category
print(df['fare'].value_counts()) #category
print(df['class'].value_counts())#category
print(df['who'].value_counts()) #category
print(df['adult_male'].value_counts()) #bool
print(df['embark_town'].value_counts()) #category
print(df['alive'].value_counts()) #Category
print(df['alone'].value_counts())#bool
print(df['embarked'].value_counts(normalize = True))
print(df.isnull().sum())



```

```python
  #data tiplerini deyish
  df['survived'] = df['survived'].astype('category')
  df['pclass'] = df['pclass'].astype('category')
  df['sex'] = df['sex'].astype('category')
  df['sibsp'] = df['sibsp'].astype('category')
  df['parch'] = df['parch'].astype('category')
  df['who'] = df['who'].astype('category')
  df['embark_town'] = df['embark_town'].astype('category')
  df['alive'] = df['alive'].astype('category')
  df['embarked'] = df['embarked'].astype('category')
  df.info()
```

```python
cat_cols = df.select_dtypes(include=["category"]).columns.tolist()
num_cols = df.select_dtypes(include=["int","float"]).columns.tolist()
```

```python
from sklearn.impute import SimpleImputer
imputer1 = SimpleImputer(strategy="most_frequent")     ##To fill Categorical Features.
imputer2 = SimpleImputer(strategy="median")
```

```python
def fill_missingno(df):
    df[cat_cols] = imputer1.fit_transform(df[cat_cols])
    df[num_cols] = imputer2.fit_transform(df[num_cols])
```

```python
fill_missingno(df)
```
