# Outlier problemi
# IQR (q3 - q1)
# Asagi outlier serheddi q1 - 1.5* IQR
# Yuxari outlier serheddi q3+ 1.5*IQR


```python
df_tip = df_1["tip"]
df_tip.describe().T
```

```python
Q1 = df_tip.quantile(0.25)
Q3 = df_tip.quantile(0.75)
IQR = Q3-Q1
```

```python
alt_hedd = Q1- 1.5*IQR
ust_hedd = Q3 + 1.5*IQR
```

```python
alt_hedd
```

```python
df_1[df_1["tip"] > ust_hedd]
```

```python
#outlier ve null deyerlerden temizlenmis datani yadda saxla
clear_df = df_1[df_1["tip"] < ust_hedd].dropna().reset_index()
```
