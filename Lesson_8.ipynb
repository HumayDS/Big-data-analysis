```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Mustafa-YAZICI/Salary-Prediction-with-ML-Models---Baseball-/main/hitters.csv')
```

```python
df.head()
```

```python
df.corr()
```

```python
df = df.drop(['Errors', 'Assists','PutOuts', 'Runs','HmRun','AtBat'],  axis = 1)
```

```python
df.head()
```

```python
#Create dummy values
dmy = pd.get_dummies(df.select_dtypes('object') ,  drop_first = True)
```

```python
dmy.head()
```

```python
#Select categoric columns
cat_columns = df.select_dtypes('object')
```

```python
#Select numeric columns
num_cols = df.drop(cat_columns , axis = 1)
```

```python
#Concatenate numeric cols with dummies
df_cleaned = pd.concat([num_cols , dmy] , axis= 1)
```

```python
df_cleaned.head()
```

```python
df_cleaned.isnull().sum()
```

```python
df_cleaned = df_cleaned.dropna()
```

```python
#Detecting out for salary with iqr method
Q1 = df_cleaned['Salary'].quantile(0.25)
Q3 = df_cleaned['Salary'].quantile(0.75)
IQR = Q3-Q1
below_bound = Q1- 2*IQR
above_bound = Q3 + 2*IQR
print(above_bound)
out_df = (df_cleaned['Salary'] > above_bound)
```

```python
#Deleting out values
df_cleaned2 = df_cleaned[~out_df]
df_cleaned2
```

```python
#Creating test train dataset
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict
```

```python
#Creating test train dataset
y = df_cleaned2['Salary']
X = df_cleaned2.drop('Salary' , axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.25,
                                                    random_state=42)
```

```python
#Building simple stats linear reg model
import statsmodels as sm
import statsmodels.regression.linear_model as sm
lm = sm.OLS(y_train, X_train)
model = lm.fit()
model.summary()
```

```python
model.predict(X)
```

```python
y
```

```python
#RMSE
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score

rmse = np.sqrt(mean_squared_error(y, model.predict(X)))
rmse
```

```python
#RMSE for test set
rmse = np.sqrt(mean_squared_error(y_test, model.predict(X_test)))
rmse
```
